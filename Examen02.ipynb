```python
def f(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 +10

def biseccion(a, b, tol=1e-12):
    if f(a) * f(b) >= 0:
        print("El intervalo [a, b] no contiene una raíz")
        return None
    c = a
    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c

intervalos = [(-4,2),(2,4),(4,6)]

for intervalo in intervalos:
    a, b = intervalo
    raiz = biseccion(a, b)
    print(f"La raíz aproximada en el intervalo [{a}, {b}] es {raiz:.4f}")
```
